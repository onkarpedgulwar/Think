```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data=pd.read_csv("/home/oslab/Desktop/akshay/1234_acc.csv")
print(data.head())
```

```python
print(data.describe())
```

```python
print(data.info())
```

```python
data_x_axis=data['x-axix']
data_against=np.arange(49)
plt.plot(data_against,data_x_axis)
plt.xlabel('time sequence')
plt.ylabel('x-axis data')
plt.title("1234 accelerometer x-axis")
plt.legend(['x-axis'])
plt.show()
```
