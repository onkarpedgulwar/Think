```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data_ten1_acc=pd.read_csv("/Users/akshaychoulwar/Desktop/Akshay/Think/ten1_acc.csv")
data_ten1_gyro=pd.read_csv("/Users/akshaychoulwar/Desktop/Akshay/Think/ten1_gyro.csv")
```

```python
print(data_ten1_gyro.describe())
```

```python
print(data_ten1_acc.describe())
```

```python
data_on_y_axis=data_ten1_acc['z-axis']
data_on_x_axis=np.arange(157)

plt.plot(data_on_x_axis,data_on_y_axis)
plt.xlabel("time sequence")
plt.ylabel("accelerometer z-axis data for 1")
plt.title("one accelerometer")
plt.show()
sns.set_style('darkgrid')
```
